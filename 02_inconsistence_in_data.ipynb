```python
import pandas as pd 
```

```python
data ={
    'date': ['2020-01-01', '2020/01/02', '01-03-2020', '2020-01-04'],
    'country': ['USA', 'U.S.A', 'America', 'United States'],
    'name': ['John Doe', 'Jane Doe', 'Jonh Doe', 'Jane Doe'],
    'sales_2020': [100, 200, None, 500],
    'sales_2021': [None, 200, 300, 500]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# standardize the date format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# Impute fill the data and try another method to fill the date format inconsistency
df['date'] = df['date'].fillna(df['date'].mode()[0])
df.head()
```

```python
country_mapping = { 'U.S.A': 'United States', 'America': 'United States', 'USA': 'United States'}
df['country'] = df['country'].replace(country_mapping)
df.head()
```

```python
# Correct the Typographical Mistakes in name
name_mapping = {'Jonh Doe': 'John Doe'} 
df['name'] = df['name'].replace(name_mapping)
df.head()
```

```python
df = df.drop_duplicates(subset="name")

df.head()
```
