# Feature Encoding

Feature Encoding is a process of converting categorical data into numerical data. This is necessary because most machine learning models require numerical input data. There are several methods to encode categorical data, such as One-Hot Encoding, Label Encoding, and Target Encoding. In this notebook, we will discuss these methods and how to implement them using Python.

- One-Hot Encoding
- Label Encoding
- Ordinal Encoding
- Binary Encoding

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# let's encode the time in labelencoder with sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
# ordinal encdoing the day olumn using specific order
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
# one hot encoding the day column
ohe = OneHotEncoder()
ohe.fit_transform(df[['sex']]).toarray()
```
